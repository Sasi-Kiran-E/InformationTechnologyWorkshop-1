```python
import pickle

"""    Make sure u have the created the 'addressBook.pickle' file
    in the same directory of this python file """

def getList():
    try:
        with open('addressBook.pickle', 'rb') as f:
            contactList = pickle.load(f)
    except EOFError:
        contactList = []
    return contactList


def pickleList(contactList):
    with open('addressBook.pickle', 'wb') as f:
        pickle.dump(contactList, f)


def createContact(name, email, phone):
    return {"name" : name, "email" : email, "phone" : phone}



def addContact(name, email, phone):
    contactList.append(createContact(name, email, phone))
    pickleList(contactList)
    print("Added to addressBook successfully")


def displayContact(lst):
    if len(lst) == 0:
        print("No contact in Adress Book")
    else:
        for contact in contactList:
            print(f'{contact["name"]} - {contact["email"]} - {contact["phone"]}')


def searchContact(lst, key):
    for contact in lst:
        if contact["name"] == key:
            print("contact found")
            print(f'{contact["name"]} - {contact["email"]} - {contact["phone"]}')
            break
    else:
        print("No contact with this name found!")


def modifyContact(lst, key):
    if len(lst) == 0:
        print("Adress Book is empty. No contact to modify")
        return
    for contact in lst:
        if contact["name"] == key:
            contact["email"] = input("Enter the new email : ")
            contact["phone"] = input("Enter the new phone : ")
            pickleList(lst)
            print("Contact modified successfully")
            print(f'Contact after modification - {contact["name"]} - {contact["email"]} - {contact["phone"]}')
            break
    else:
        print("contact not found in the Address Book")


def deleteContact(lst, key):
    if len(lst) == 0:
        print("Adress Book is empty. No contact to delete")
        return
    for contact in lst:
        if contact["name"] == key:
            lst.remove(contact)
            print("successfully deleted the contact")
            pickleList(lst)
            break
    else:
        print("No contact with this name")


contactList = getList()

while True:

    print('-+-' * 15)
    print('What u wanna do ?')
    print('1 - Add contact')
    print('2 - Display contact')
    print('3 - Search contact')
    print('4 - Modify contact')
    print('5 - Delete contact')
    print('0 - exit')
    print('-+-' * 15)

    choice = input()

    if choice == '1':
        name = input("Enter the contact's name - ")
        email = input("Enter the contact's email id - ")
        phone = input("Enter the contact's phone - ")
        addContact(name, email, phone)

    elif choice == '2':
        displayContact(contactList)

    elif choice == '3':
        name = input("Enter the contact's name that u wanna search for : ")
        searchContact(contactList, name)

    elif choice == '4':
        name = input("Enter the contact's name whose info u wanna modify : ")
        modifyContact(contactList, name)

    elif choice == '5':
        name = input("Enter the contact's name whom u wanna delete : ")
        deleteContact(contactList, name)

    elif choice == '0':
        print("Exited successfully")
        break
        
    else:
        print("Enter valid choice !")
```
