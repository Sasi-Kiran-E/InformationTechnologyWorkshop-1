```python

foodMenu = {1 : 'Chicken Strips', 2 : 'French Fries', 3 : 'Hamburger', 4 : 'Hotdog', 
            5 : 'Large Drink', 6 : 'Medium Drink', 7 : 'Milk Shake', 8 : 'Salad',
            9 : 'Small Drink'
            }

costMenu = {1 : 3.50, 2 : 2.50, 3 : 4.00, 4 : 3.50, 5 : 1.75, 6 : 1.5, 7 : 2.25, 8 : 3.75, 9 : 1.25}


class Restaurant:

    def __init__(self):
        self.order = {}
        self.totalCost = 0
        for key in range(1, 10):
            self.order[key] = 0

    def calculate(self):
        for i in range(1, 10):
            self.totalCost += self.order[i] * costMenu[i]

    def takeOrder(self):
        string = input("Give the order : ")
        for c in string:
            k = ord(c) - ord('0')
            self.order[k] += 1
        self.calculate()

    def displayOrder(self):
        for i in range(1, 10):
            if self.order[i] == 0:
                continue
            print(f'{foodMenu[i]} - {self.order[i]}')
        print(f'Total cost - ${self.totalCost}')
        print('-' * 30)



def displayFoodMenu(foodMenu, costMenu):
    print('-+-' * 10)
    for i in range(1, 10):
        print(f'{i}. {foodMenu[i]} - ${costMenu[i]}')
    print('-+-' * 10)


customer = Restaurant()

displayFoodMenu(foodMenu, costMenu)

while True:

    print("Enter ur choice : ")
    print('yes - to continue order')
    print('no - place order and exit')
    print('menu - to see items and cost')
    print('-' * 30)

    choice = input()

    if choice == 'menu':
        displayFoodMenu(foodMenu, costMenu)

    elif choice == 'yes':
        customer.takeOrder()
        print("you ordered -> ")
        customer.displayOrder()

    else:
        print('Hope U enjoyed !!!')
        break
```
